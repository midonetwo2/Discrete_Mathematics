# 1. Proving Identities

## 1.1. $A \setminus B = A \cap B^C$
**Proof:**

By the definition of complement, we have that $B^C = \mathcal{U} - B$, we intersect both sides with the set $A$ and get:
$$A \cap B^C = A \cap \mathcal{U} - A \cap B$$

By the Domination law, $A \cap \mathcal{U} = A$, thus we have:
$$A \cap B^C = A - A \cap B$$

By definition, $A \setminus B = A - A \cap B$, which implies:
$$A \setminus B = A \cap B^C$$

**Q.E.D**

## 1.2. $B = (B \cap A) \cup (B \cap A^C)$
**Proof:**

By the Distributivity law, we have:
$$(B \cap A) \cup (B \cap A^C) = B \cap (A \cup A^C)$$

Since $A \cup A^C = \mathcal{U}$ (by the complement law), we get:
$$B \cap \mathcal{U} = B$$

Thus, we have:
$$B = (B \cap A) \cup (B \cap A^C)$$

**Q.E.D**

## 1.3. $B \setminus A = B \cap A^C$
**Proof:**

By the definition of complement, we have $A^C = \mathcal{U} - A$, and intersecting both sides with the set $B$ gives:
$$B \cap A^C = B \cap \mathcal{U} - B \cap A$$

Using the Domination law, $B \cap \mathcal{U} = B$, so we have:
$$B \cap A^C = B - B \cap A$$

By definition, $B \setminus A = B - B \cap A$, which implies:
$$B \setminus A = B \cap A^C$$

**Q.E.D**

## 1.4. $A \cup B = A \cup (B \cap A^C)$
**Proof:**

Using the Distributivity law, we have:
$$A \cup (B \cap A^C) = (A \cup B) \cap (A \cup A^C)$$

Since $A \cup A^C = \mathcal{U}$ (by the complement law), we get:
$$(A \cup B) \cap \mathcal{U} = A \cup B$$

Thus, we have:
$$A \cup B = A \cup (B \cap A^C)$$

**Q.E.D**

## 1.5. $(\bigcup_{i = 1}^{\infty} A^C_i)^C = \bigcap_{i = 1}^{\infty}A_i$
**Proof:**

By De Morgan's Law, we distribute the complement inside the parenthesis:
$$(\bigcap_{i = 1}^{\infty} A_i^C)^C = \bigcap_{i = 1}^{\infty} A_i$$

**Q.E.D**

## 1.6. $|A \cup B| = |A| + |B| - |A \cap B|$
**Proof:**

**Case 1: Disjoint Sets**
If $A$ and $B$ are disjoint, their intersection is the empty set: $|A \cap B| = 0$. The cardinality of the union of the sets is given by: $|A \cup B| = |A| + |B|$.

**Case 2: Non-disjoint Sets**
For non-disjoint sets $A$ and $B$, we count the size of both sets. Since we double count the common elements in $A \cap B$, we subtract $|A \cap B|$. This gives: $|A \cup B| = |A| + |B| - |A \cap B|$.

**Q.E.D**
